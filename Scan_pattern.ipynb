```python
stock = []
for j in url_df['url']:
    source = requests.get(j).text
    soup = BeautifulSoup(source, 'html.parser')       
    table_rows = soup.find_all("tr")
    for tr in table_rows:
        td = tr.find_all("td")
        row = [tr.text for tr in td]
        stock.append(row)
```
